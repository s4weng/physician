```python
df = pd.read_csv("./physician2.csv")

df["topics_list"] = df["what_topics_is_this_review_about"].str.split("\r")

topics = df["topics_list"].dropna()

import itertools
attributes = pd.DataFrame(topics.apply(lambda x: dict(zip(x, itertools.repeat(1)))).tolist(), index=topics.index)

df = df.join(attributes)

#test data
df_test = pd.read_csv("./physicians_test.csv")

df_test["topics_list"] = df_test["what_topics_is_this_review_about"].str.split("\r")

topics_test = df["topics_list"].dropna()
attributes_test = pd.DataFrame(topics.apply(lambda x: dict(zip(x, itertools.repeat(1)))).tolist(), index=topics_test.index)

df_test = df_test.join(attributes_test)

df = df.append(df_test)

df = df[df.what_topics_is_this_review_about.notnull()]
df.fillna(0, inplace=True)

from sklearn.cross_validation import train_test_split

# from sklearn.preprocessing import MultiLabelBinarizer
binarizer = MultiLabelBinarizer()

test_target = binarizer.fit_transform(df.topics_list)

#df.dropna(inplace=True)
X_train, X_test, y_train, y_test = train_test_split(df.comment, df.topics_list, test_size=0.25, random_state=42)

test_target
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

df = df[pd.notnull(df['topics_list'])]

text_clf = Pipeline([('vect', CountVectorizer(stop_words="english")),
                      ('tfidf', TfidfTransformer()),
                      ('clf', OneVsRestClassifier(RandomForestClassifier()))
 ])

from sklearn.grid_search import GridSearchCV
parameters = {'vect__ngram_range': [(1,1),(1, 2),(1,3)],
               'tfidf__use_idf': (True, False),
              'tfidf__norm': ('l1', 'l2'),
              'tfidf__smooth_idf': (True, False),
              'tfidf__sublinear_tf': (True, False),
              'vect__min_df': (.05,.1),
              'vect__max_df': (0.5, 0.75, 1.0),
            'vect__max_features': (None, 5000, 10000, 50000),
}
gs_clf = GridSearchCV(text_clf, parameters, cv=5, n_jobs=-1, verbose=1)

train_target = binarizer.transform(y_train)

#text_clf = gs_clf.fit(train_data, train_target)
gs_clf.fit(X_train, train_target)
```

```python
#let's get some final scores broken down by label
from sklearn.metrics import precision_recall_fscore_support
predicted = gs_clf.predict(df.comment)
precision_recall_fscore_support(train_target, predicted, average=None)

from sklearn.metrics import classification_report
print(classification_report(train_target, predicted
, target_names=list(binarizer.classes_)))
```
